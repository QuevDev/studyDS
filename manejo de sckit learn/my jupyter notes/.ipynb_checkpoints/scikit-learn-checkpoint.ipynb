<h2>Como aprenden las maquinas?</h2><br>
<h3>Las maquinas aprenden de los datos disponibles.</h3><br>
<p>Desde el punto de vista de los datos podemos aplicar tres tecnicas segun la naturaleza y disponibilidad de los mismos.</p><br>
<li>Aprendizaje supervisado.</li>
<li>Aprendizaje no supervisado.</li>
<li>Aprendizaje por refuerzo</li><br>

<h3>Aprendizaje supervisado(Por observacion.)</h3><br>
<p>Si de nuestro conjunto de datos podemos extraer con anticipacion informacion precisa del resultado que esperamos.</p>
<br>
<h3>Aprendizaje Por refuerzo(Prueba y error.)</h3><br>
<p>Si no tenemos informacion precisa sobre lo que esperamos, pero si podemos evaluar si una decision tomada por la maquina es buena o mala.</p>
<br>
<h3>Aprendizaje no supervisado(Por descrubrimiento.)</h3><br>
<p>Finalmente, si no sabemos que esperar de nuestros datos y queremos explorar, la estructura o relaciones de nuestro dataset.</p>

<h3>Como afectan los features a los modelos de ML</h3>
<br>
<p>¿Qué son los features?,son atributos de nuestros modelo que usamos para realizar una interferencia o prediccion. Son las veriables de entrada.</p><br>
<b>Mas features siempre es mejor, verdad?</b><br>
<p>la respuesta correcta es : NO <p><br>
En realidad si tenemos variables que son irrelevantes pasarean estas cosas: <br>
    <li>Se abrira paso al ruido</li>
    <li>Aumenta el costo computacional</li>
    <li>Si introduciomos demasiados features y estos tienen valores faltantes, se haran sesgos muy significativos y vamos a perder esa capacidad de prediccion.</li><br>
   <br>
    <b>Una de las formas de saber que nuestyros features han sido bien seleccionados es con el sesgo y la varianza.</b><br>
   <la> <li>Una mala seleccion de nuestro features nos puede llevar a alguno de esos dos escenarios indeseados.</li> </la> 
    <br>
      <b>Algo que debemos recordar es que nuestro modelo de ML puede caer en unos de 2 escenarios que debemos evitar :</b><br>
      <li><b>Underffiting</b>: significa que nuestro modelo es demasiado simple, en donde nuestro modelo no esta captando los featues y nuestras variable de salida, por lo cual debemos investigar variables con mas signbificados o combinciones o transformaciones para poder llegar a nuestra variable de salida.</li>
       <la><li><b>Overfitting</b>: Siginifica que nuestro modelo es demasiado completo y nuestro algoritmos va a intentar ajustarse a los datos que tenemos, pero no se va comportar bien con los datos del mundo real, si tenemos overfitting lo mejor es intentar seleccionar los features de una manera mas critica descartando aquellos que no aporten.</li></la> <img src='https://static.platzi.com/media/user_upload/2-9e725af1-31e2-4381-864b-77072a2546c2.jpg' />
    <br>
    <b>Que podemos hacer para solucionar estos problemas?</b>
    <la>
        <li>Aplicar tecnicas reduccion de la dimensionalidad, utilizaremos el algoritmo de PCA</li>
        <li>Aplicar la tecnica de la regulacion, que consiste en penalizar auqellos features que no le esten aportando o que le esten restando informacion a nuestro modelo.</li>
        <li>Balanceo : Se utilizara Oversampling y undersampling en porblemas de rendimiento donde tengamos un conjunto de datos que esta desblanaceado, por ejemplo un problema de clasificacion donde tenemos muchos ejemplos de una categoria y muy pocos de otra.</li>
    </la>

<h1>PCA(Principal component analysis)</h1>
<br>
<b>Por que usariamos este algoritmo?</b><br>
<la>
    <li>Porque en ML es normal encontrarnos con problemas donde tengamo una enorme cantidad de features en donde hay relaciones complejas entre ellos y con la variable que queramos predecir.</li>
</la> 
<p>Pistas donde se puede utilizar este algoritmo</p>
<la>
    <li>Nuestro dataset tiene un numero alto de features y no todos sean significativos</li>
    <li>Hay una alta correlacion entre los features</li>
    <li>cuando hay overfiting</li>
    <li>Cuando implica un alto coste computacional</li>
</la> 
<br>
<b>En que consiste el algoritmo PCA</b>
<p>Basicamente en reducir la complejidad del problema.</p>
<la>
    <li>Seleccionando solamentre las variables relevantes</li>
    <li>Combinandolas en nuevas variablesque mantengan la informacion mas importante(varianza de los features)</li>
    <img src="https://static.platzi.com/media/user_upload/2-407a2b92-0772-4cbc-9f45-6d91fb33dc12.jpg"/>
</la>

<b>Pasos para llevar a cabo el algoritmo PCA</b>
<la>
    <li>Primero se calcula la matriz de covarianza para expresar las relaciones entre nuestros features.</li>
    <li>Hallamos los vectores propios y valores propios con mayor variabilidad, esto es, aporte de mas informacion.</li>
</la>


<h1 style="color : red;">Preparacion de datos para PCA</h1>

```python
import pandas as pd 
import matplotlib.pyplot as plt
import sklearn 

from sklearn.decomposition import PCA
from sklearn.decomposition import IncrementalPCA

from sklearn.linear_model import LogisticRegression

#nos ayuda a normalizar nuestros datos para que todos esten en la escala de 0 a 1
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

#implementacion de KERNEL
from sklearn.decomposition import KernelPCA


#carga del dataset 
dt_heart = pd.read_csv('./Downloads/heart.csv')
dt_heart.head(5)
```

```python
#guardar features
dt_features = dt_heart.drop(['target'],axis=1)
dt_target = dt_heart['target']

#cargar los datos, ajustar nuestro modelo y aplicar la transformacion
dt_features = StandardScaler().fit_transform(dt_features)

#partir el conjunto de entrenamiento
X_train,X_test,Y_train,Y_test = train_test_split(dt_features,dt_target,test_size=0.4,random_state=42)
```

```python
#n_componenbts = min(n_muestras,n_features)
pca = PCA(n_components=4)
pca.fit(X_train)
```

```python
#medir la varianza
plt.plot(range(len(pca.explained_variance_)),pca.explained_variance_ratio_)
plt.show()
```

```python
logistic = LogisticRegression(solver='lbfgs')
```

```python
dt_train = pca.transform(X_train)
dt_test = pca.transform(X_test)

logistic.fit(dt_train,Y_train)

print(f' SCORE : {logistic.score(dt_test,Y_test)}')
```

<h1 style='color : red;'>Implementacion de Lasso y Ridge</h1>
<p><b>Se utilizara el dataset de felicidad</b></p>

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import sklearn

from sklearn.linear_model import  LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
dt_felicidad = pd.read_csv('./Downloads/felicidad.csv')
dt_felicidad.head(5)
```

```python
import seaborn as sb

sb.heatmap(dt_felicidad.corr(),annot=True)
```

```python
#Seprar nuestro dataset en features y targets
dataset = dt_felicidad.copy()

#Features
X = dataset[['gdp','family','lifexp','freedom','corruption','generosity','dystopia']]
y = dataset[['score']]

X.shape,y.shape

#seprar en training y test / test size separa 25% para test y un 75% para train
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)

#lineal 
modelLineal = LinearRegression().fit(X_train,y_train)
y_predict_linear = modelLineal.predict(X_test)

#model lasso
modelLasso = Lasso(alpha=0.02).fit(X_train,y_train)
y_predict_lasso = modelLasso.predict(X_test)

#Ridge 
modelRidge = Ridge(alpha=0.02).fit(X_train,y_train)
y_predict_ridge = modelRidge.predict(X_test)

#calculo de las perdidas por el metodo de error minimo cuadratico
linear_loss = mean_squared_error(y_test,y_predict_linear)

lasso_loss = mean_squared_error(y_test,y_predict_lasso)

ridge_loss = mean_squared_error(y_test,y_predict_ridge)

#imprimir errores

print(f'Linear loss : {linear_loss}')
print(f'Lasso loss : {lasso_loss}')
print(f'Ridge loss : {ridge_loss}')
```

```python
#coeficiones
#lasso
print(f'coef lasso: {modelLasso.coef_}')

print('-'*32)

#ridge
print(f'coef ridge : {modelRidge.coef_}')

print('-'*32)
#lineal
print(f'coef lineal : {modelLineal.coef_}')

```

<h1>Valores atipicos</h1>
<br>
<li>Un valor atpico es cualquier medicion que se encuentre por fuera del comportamiento general de una muestra de datos.</li>
<li>Pueden indicar variabilidad, errores de medicion, o novedades.</li>
<br>
<b>Por que son problematicos los valores atipicos</b>
<li>Pueden generar sesgos importantes en los modelos de ML.</li>
<li>a veces contienen informacion relevante sobre la naturaleza de los datos.</li>
<li>Deteccion temprana de fallos.</li>
<br>
<b>Como identificarlos?</b>
<li>Z-Score : mide la ditancia(En desviaciones estandar) de x punto dado a la media</li>
<li>Tecnicas de clustering como <b>DBSCAN</b></li>
<li>Si q 'menor que' <span style='color : blue;'>Q1-1.5*IQR</span> O q 'mayor que' <span style='color : blue;'>Q3+1.5*IQR</span> </li>

<h1>Regresiones robustas</h1>
<p>Sci-kit nos ofrece algunos modelos especificos para abordar el problema de los valores atipicos.</p><br>
<li><b>RANSAC</b></li>
<li><b>Huber Reggressor</b></li>
<br>
<h3>RANSAC</h3>
<p><b>Random Sample Consensus</b></p>

<p>Usamos una muestra aleatoria sobre el conjunto de datos que tenemos, buscamos la muestra que mas datos "buenos" logre incluir</p>
<p>El modelo asume que los "malos errores" no tienen patrones especificos</p>

<h3>Huber Reggresor</h3>
<p>No ignora los valores atipicos, disminuye su influencia en el modelo</p>
<p>Los datos son tratados como atipicos si el error absoluto de nustra perdida esta por encima de un umbral llamado epsilon</p>
<p>Se ha demostrado que un valor de <span style='color : blue;'>epsilon = 1.35</span> logras un 95% de eficiencia estadistica.</p>


<h1 style='color : red;'>Regresion robusta</h1>

```python
import pandas as pd 
from sklearn.linear_model import (
    RANSACRegressor, HuberRegressor
)

from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

import matplotlib.pyplot as plt

import warnings
warnings.simplefilter('ignore')
```

```python
dataset = pd.read_csv('./Downloads/felicidad.csv')
dataset.head(5)
```

```python
#definir features
X = dataset.drop(['country','score'],axis=1)
#definir target
y = dataset[['score']]
```

```python
#crear variables de train y test 
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
#como trabajamos con varios estimadores haremos algo mas profesional
estimadores = {'SVR' : SVR(gamma='auto',C=1.0,epsilon=0.1),
               'RANSAC' : RANSACRegressor(),
               'HUBER' : HuberRegressor(epsilon=1.35)}
```

```python
for name,estimador in estimadores.items():
    estimador.fit(X_train,y_train)
    predictions = estimador.predict(X_test)
    
    print(name)
    print(f'MSE : {mean_squared_error(y_test,predictions)}')
    print('===='*10)
    plt.scatter(y_test,predictions)
    plt.plot(predictions,predictions,color='r')
    plt.ylabel('Prediccion')
    plt.xlabel('Score real')
    plt.show()
```

<h2>Baggin</h2>
<p>1) Creamos particiones aleatorias(Uniforme y con remplazo) del conjunto de datos original.</p>
<p>2) Se construyen modelos de aprendizaje para cada uno de los subconjuntos por aparte.</p>
<p>3) La respuesta final es la combinacion(por ej. por votaciuon entre las respuestas individuales.)</p>
<br>
<p><b>Modelos Emsabnlados basados en Bagging : </b></p>
<li>Random Forest</li>
<li>Voting Classifiers/Regressors</li>
<li>En general se puede aplicar sobre cualquier familia de modelos de ML.</li>
<br>
<h2>Boosting</h2>
<p>Impulsar/propulsar</p>
<p>Busca fortalecer gradualmente un modelo de aprendizaje usando siempre el error residual de las etapas anteriores.</p>
<p>El resultado final tambien se consigue por conseso entre todos los modelos.</p>

<li>Adaboost</li>

<h1>Impementacion de metodos de emsable</h1>

```python
import pandas as pd

from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import BaggingClassifier

from sklearn.ensemble import GradientBoostingClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
dt_heart = pd.read_csv('./Downloads/heart.csv')
dt_heart.head(5)
```

```python
#definir features y target 
X = dt_heart.drop(['target'],axis=1)
y = dt_heart['target']
```

```python
#creacion de variables train y test
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.35)
```

```python
knn = KNeighborsClassifier().fit(X_train,y_train)
knn_predict = knn.predict(X_test)
print('='*20)
print(f'accuracy : {accuracy_score(knn_predict,y_test)}')
```

```python
#creacion del emsable 
bag_class = BaggingClassifier(base_estimator=KNeighborsClassifier(),n_estimators=50).fit(X_train,y_train)
bag_predict = bag_class.predict(X_test)
print('='*20)
print(f'accuracy : {accuracy_score(bag_predict,y_test)}')
```

<h1>Tarea : comparar con distintos estimadores Bagging</h1>

```python
#boosting

#n_estimaros es cuantos arboles quiero
boost = GradientBoostingClassifier(n_estimators=50).fit(X_train,y_train)
boost_pred = boost.predict(X_test)

#Gradient boosting crear arboles peque;os y los entrena

print('='*20)
print(f'accuracy : {accuracy_score(boost_pred,y_test)}')
```

<b>Problema</b>: Parece que encontramos un modelo de aprendizaje que parece funcionar, pero esto puede implicar que ahora tenemos que encontrar la optimización de cada uno de los parámetros de este modelo, encontrar el que mejor se ajuste y el que mejor resultado nos de.

Es facil perderse entre los conceptos de tantos parámetros. Tenemos flexibilidad para algoritmos básicos de Machine Learning, pero facil perderse.
Es difícil medir la sensibilidad de los mismos manualmente.
Es COSTOSO, en tiempo humano y computacionalmente.
Scikit Learn nos ofrece enfoques para automatizar el proceso de optimización paramétrica. Existen 3 enfoques principales, estos son:

<b>Optimización manual</b>
    <li>Escoger el modelo que queremos ajustar</li>
    <li>Buscar en la documentacion de scikit learn</li>
    <li>Identificar los p;osibles ajustes</li>
    <li>Probar combinaciones una por una iterando a traves de listas</li>

<b>Optimizacion por grilla de parámetros | GridSearchCV</b>
<p>Una forma organizada,exhaustiva y sistematica de probar todos los parametros que le digamos que tenga que probar, con los respectivosrangos de valores que le aportaremos</p>
<li>Definir una o varias metricas que queramos optimizar</li>
<li>Identificar losposibles valores que pueden tener los parametros</li>
<li>Crear un diccionario de parametros</li>
<li>Usar CV</li>
<li>Entrenar el modelo</li>
<p>La grilla de parametrosnos define grupos de parametros que sera probadosa en todas sus combinaciones(un grupo a la vez).</p>

<b>Optimizacion por búsqueda aleatorizada | RandomizedSearchCV</b>
<p>Si no tenemos tanto tiempo para una prueba exhaustiva o queremos combinaciones alaeatorias usaremos este modelo.Lo unico diferente es que no busca de forma exhausitva si no de forma aleatoria, selecciona las mejores combinaciones que se hicieron. En este metodo, definicimos las escalas de valores para cada uno de los parametros seleccionados, el sistema probara varias iteraciones(Configurables segun los recursos) y mostrara la mejor combinacion encontrada : </p>

<h2>GridSearchCV vs RandomizedSearchCV</h2>
<b>GridSearchCV</b>
<li>Cuando se quiera realizar un estrudio a fondo sobre las implicaciones de los parametros</li>
<li>Se tenga tiempo</li>
<li>Se tenga el poder de procesamiento</li>
<b>RandomizedSearchCV</b>
<li>Cuando se quiera explorar posibles optmizadores</li>

```python
import pandas as pd 
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
```

```python
dt = pd.read_csv('./Downloads/felicidad.csv')
dt.head()
```

```python
# 
reg = RandomForestRegressor()
parametros = {
    'n_estimators' : range(4,16),
    'criterion' : ['mse','mae'],
    'max_depth' : range(2,11) #define profundidad del arbol
}
```

```python
X = dt.drop(['country','rank','score'],axis=1)
y = dt[['score']]

rand_est = RandomizedSearchCV(reg,parametros,n_iter=10,cv=3,scoring='neg_mean_absolute_error').fit(X,y) #va tomar 10 diferentes configuraciones
#cv = 3 va partir nuestro ran en 3 pliegues
```

```python
rand_est.best_estimator_ #configuracion adecuada
```

```python
rand_est.best_params_
```

<br>

Cada clase será un archivo que tenga su propia responsabilidad y se encargue específicamente de una tareas concreta.
